```python
####최종완성본###

import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
import os
import cv2
import numpy as np
from sklearn.metrics import precision_score, recall_score
import tensorflow as tf
import tensorflow_hub as hub
#이미지누락처리
from PIL import ImageFile, Image
ImageFile.LOAD_TRUNCATED_IMAGES = True



test_data_dir = r'C:\Users\sj990\Desktop\테스트이미지들\테스트할것들2'
class_labels = os.listdir(test_data_dir)  # List of class labels based on folder names

test_images = []  # List to store test images
ground_truth_labels = []  # List to store ground truth labels

for label_idx, label in enumerate(class_labels):
    label_dir = os.path.join(test_data_dir, label)
    image_files = os.listdir(label_dir)
    
    for image_file in image_files:
        image_path = os.path.join(label_dir, image_file)
        
        try:
            img = Image.open(image_path)
            img = img.resize((256, 256))  # Resize the image
            img = img_to_array(img)
            img = img.astype('float32') / 255.0
            test_images.append(img)
            ground_truth_labels.append(label_idx)  # Use class index as label
        except (IOError, OSError) as e:
            print(f"Skipping {image_path} due to an error: {e}")

test_images = np.array(test_images)
ground_truth_labels = np.array(ground_truth_labels)

model_path = r'C:\Users\sj990\Desktop\9.23_test\my_model_all_v2.h5'
model_path2 = r'C:\Users\sj990\Desktop\9.23_test\my_model_efficientnetb0.h5'

custom_model = tf.keras.models.load_model(model_path2,  custom_objects={'KerasLayer': hub.KerasLayer})

predicted_probs = custom_model.predict(test_images)
predicted_labels = np.argmax(predicted_probs, axis=1)
```

```python
#예측 틀린거 실제이미지

incorrect_indices = np.where(predicted_labels != ground_truth_labels)[0]

# You can print the indices of incorrect predictions
print("Indices of incorrect predictions:", incorrect_indices)

# You can also display some of the incorrectly predicted images
import matplotlib.pyplot as plt

num_incorrect_to_display = 5  # You can change this to display more or fewer images
for i in range(min(num_incorrect_to_display, len(incorrect_indices))):
    index = incorrect_indices[i]
    predicted_label = class_labels[predicted_labels[index]]
    ground_truth_label = class_labels[ground_truth_labels[index]]
    
    img = test_images[index]
    img = (img * 255).astype(np.uint8)  # Convert the image back to uint8 for displaying
    
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.title(f"Predicted: {predicted_label}\nGround Truth: {ground_truth_label}")
    plt.axis('off')
    plt.show()
    
#성능평가
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix, accuracy_score

# Calculate precision, recall, and F1 score
accuracy = accuracy_score(ground_truth_labels, predicted_labels)
precision = precision_score(ground_truth_labels, predicted_labels, average='weighted')
recall = recall_score(ground_truth_labels, predicted_labels, average='weighted')
f1 = f1_score(ground_truth_labels, predicted_labels, average='weighted')

print(f'Accuracy: {accuracy * 100:f}%')
print(f'Precision: {precision * 100:f}%')
print(f'Recall: {recall * 100:f}%')
print(f'F1 Score: {f1 * 100:f}%\n\n')
```
